```python
calendar = pd.read_csv(r"Vic_GTFS_data\metropolitan\calendar.txt")
```

```python
melcen_stop_time = stop_time.loc[stop_time['stop_id'] == 19842]
```

```python
sunshine_stop_time
```

```python
com_stop = pd.merge(sunshine_stop_time, melcen_stop_time, on="trip_id", how="inner")
```

```python
com_stop = com_stop[['trip_id', 'departure_time_x', "arrival_time_y"]]
```

```python
def fix_24(time):
    if time[0:2] == "24":
        time = "00" + time[2:]
        return time
    elif time[0:2] == '25':
        time = "01" + time[2:]
        return time
    else:
        return time
```

```python
com_stop['departure_time_x'] = com_stop['departure_time_x'].apply(fix_24)
```

```python
com_stop['arrival_time_y'] = com_stop['arrival_time_y'].apply(fix_24)
```

```python
com_stop['departure_time_x'] = com_stop['departure_time_x'].apply(lambda x: datetime.strptime(x, "%H:%M:%S").time())
```

```python
com_stop['arrival_time_y'] = com_stop['arrival_time_y'].apply(lambda x: datetime.strptime(x, "%H:%M:%S").time())
```

```python
# time intervalg
start =  datetime.strptime("07:00:00","%H:%M:%S").time()
end = datetime.strptime("09:00:00","%H:%M:%S").time()
```

```python
# filter row that not in interval
# com_stop = com_stop.loc[(com_stop['departure_time_x'] >= start) & (com_stop['arrival_time_y'] <= end)]
com_stop = com_stop.loc[com_stop['departure_time_x'].between(start, end)]
com_stop = com_stop.loc[com_stop['arrival_time_y'].between(start, end)]
```

```python
# keep only the trip haeding to mc
com_stop = com_stop.loc[com_stop['arrival_time_y'] > com_stop['departure_time_x']]
```

```python
com_stop.info()
```

```python
com_stop = pd.merge(com_stop, trip_df, how="left", on="trip_id")
```

```python
com_stop = com_stop[['trip_id', 'departure_time_x', 'arrival_time_y', 'route_id', 'service_id', 'direction_id']]
```

```python
com_stop.info()
```

```python
com_stop = com_stop.loc[com_stop['direction_id'] == 0]
```

```python
com_stop = pd.merge(com_stop, calendar.drop(['start_date', 'end_date'], axis=1), how='left', on= 'service_id')
```

```python
com_stop = com_stop.drop(['saturday', 'sunday'], axis=1)
```

```python
com_stop = com_stop.loc[(com_stop['monday'] == 1)&(com_stop['tuesday'] == 1)&(com_stop['wednesday'] == 1)&(com_stop['thursday'] == 1)&(com_stop['friday'] == 1) ]
```

```python
route_df = route_df[['route_id', 'route_type']]
```

```python
com_stop = pd.merge(com_stop, route_df, how='left', on='route_id')
```

```python
com_stop
```

```python
# change time object to time delta
com_stop['departure_time_x_min'] = com_stop['departure_time_x'].apply(lambda x: timedelta(hours=x.hour, minutes=x.minute, seconds=x.second))
com_stop['arrival_time_y_min'] = com_stop['arrival_time_y'].apply(lambda x: timedelta(hours=x.hour, minutes=x.minute, seconds=x.second))
```

```python
com_stop
```

```python
com_stop['time_diff'] = abs(com_stop['departure_time_x_min'] - com_stop['arrival_time_y_min'])
com_stop['time_diff'] = com_stop['time_diff'].apply(lambda x: x.total_seconds()/60)
```
